```python
import numpy as np
import matplotlib.pyplot as plt

#k in 1/Mpc
def tk_eh_func(k,Ob0=0.2,Om0=0.2,h=0.5,Tcmb0=2.725):
    Obh2 = Ob0 * h ** 2
    Omh2 = Om0 * h ** 2
    f_baryon = Ob0 / Om0

    theta_cmb = Tcmb0 / 2.7

    z_eq = 2.5e4 * Omh2 * theta_cmb ** (-4)  # really 1+z
    k_eq = 7.46e-2 * Omh2 * theta_cmb ** (-2)  # units Mpc^-1 (no h!)

    z_drag_b1 = (
        0.313 * Omh2 ** -0.419 * (1.0 + 0.607 * Omh2 ** 0.674)
    )
    z_drag_b2 = 0.238 * Omh2 ** 0.223
    z_drag = (
        1291.0
        * Omh2 ** 0.251
        / (1.0 + 0.659 * Omh2 ** 0.828)
        * (1.0 + z_drag_b1 * Obh2 ** z_drag_b2)
    )

    r_drag = (
        31.5 * Obh2 * theta_cmb ** -4 * (1000.0 / (1 + z_drag))
    )
    r_eq = 31.5 * Obh2 * theta_cmb ** -4 * (1000.0 / z_eq)

    sound_horizon = (
        (2.0 / (3.0 * k_eq))
        * np.sqrt(6.0 / r_eq)
        * np.log(
            (np.sqrt(1.0 + r_drag) + np.sqrt(r_drag + r_eq))
            / (1.0 + np.sqrt(r_eq))
        )
    )

    k_silk = (
        1.6
        * Obh2 ** 0.52
        * Omh2 ** 0.73
        * (1.0 + (10.4 * Omh2) ** (-0.95))
    )

    alpha_c_a1 = (46.9 * Omh2) ** 0.670 * (
        1.0 + (32.1 * Omh2) ** (-0.532)
    )
    alpha_c_a2 = (12.0 * Omh2) ** 0.424 * (
        1.0 + (45.0 * Omh2) ** (-0.582)
    )
    alpha_c = alpha_c_a1 ** (-f_baryon) * alpha_c_a2 ** (
        -f_baryon ** 3
    )

    beta_c_b1 = 0.944 / (1.0 + (458.0 * Omh2) ** -0.708)
    beta_c_b2 = (0.395 * Omh2) ** -0.0266
    beta_c = 1.0 / (1.0 + beta_c_b1 * ((1 - f_baryon) ** beta_c_b2 - 1))

    y = z_eq / (1 + z_drag)
    alpha_b_G = y * (
        -6 * np.sqrt(1 + y)
        + (2 + 3 * y) * np.log((np.sqrt(1 + y) + 1) / (np.sqrt(1 + y) - 1))
    )
    alpha_b = (
        2.07
        * k_eq
        * sound_horizon
        * (1.0 + r_drag) ** -0.75
        * alpha_b_G
    )

    beta_node = 8.41 * Omh2 ** 0.435
    beta_b = (
        0.5
        + f_baryon
        + (3.0 - 2.0 * f_baryon) * np.sqrt((17.2 * Omh2) ** 2 + 1.0)
    )


    # Get k in Mpc^-1
    k = k

    q = k / (13.41 * k_eq)
    ks = k * sound_horizon

    T_c_ln_beta = np.log(np.e + 1.8 * beta_c * q)
    T_c_ln_nobeta = np.log(np.e + 1.8 * q)
    T_c_C_alpha = (14.2 / alpha_c) + 386.0 / (1.0 + 69.9 * q ** 1.08)
    T_c_C_noalpha = 14.2 + 386.0 / (1.0 + 69.9 * q ** 1.08)

    T_c_f = 1.0 / (1.0 + (ks / 5.4) ** 4)

    def term(a, b):
        return a / (a + b * q ** 2)

    T_c = T_c_f * term(T_c_ln_beta, T_c_C_noalpha) + (1 - T_c_f) * term(
        T_c_ln_beta, T_c_C_alpha
    )

    s_tilde = sound_horizon / (1.0 + (beta_node / ks) ** 3) ** (1.0 / 3.0)
    ks_tilde = k * s_tilde

    T_b_T0 = term(T_c_ln_nobeta, T_c_C_noalpha)
    Tb1 = T_b_T0 / (1.0 + (ks / 5.2) ** 2)
    Tb2 = (alpha_b / (1.0 + (beta_b / ks) ** 3)) * np.exp(
        -((k / k_silk) ** 1.4)
    )
    T_b = np.sin(ks_tilde) / ks_tilde * (Tb1 + Tb2)

    return f_baryon * T_b + (1 - f_baryon) * T_c

def tk_bbks_func(k,Ob0=0.2,Om0=0.2,h=0.5,Tcmb0=2.725):
    tfac = Tcmb0 / 2.7
    q = tfac * tfac * k / (Om0 * h * h *np.exp( -Ob0*(1.+(2.*h)**0.5/Om0) ) )
    tk2= (np.log(1. + 2.34*q) / (2.34*q))** 2.0 /(1. + 3.89*q + (16.1*q)**2.0 + (5.46*q)**3.0 + (6.71*q)**4.0)**0.5
    return np.sqrt(tk2)
```

```python
def GISDB(m,k,z):
    h=0.7
    Onuh2_mass=m/93.14
    Ocdh2=0.14
    Omh2 = Onuh2_mass+Ocdh2
    f_nu=Onuh2_mass/(Onuh2_mass+Ocdh2)

    alpha = 4.0 
    Delta_LCDM = 4.8e-3 
    theta_cmb = 2.7255/2.7
    k_eq = 0.0746*Omh2/theta_cmb**2.
    #k_eq = 0.015 * h
    R_LCDM = 1.0 + Delta_LCDM * np.tanh(alpha*k/k_eq)
    if(m==0.):
        return R_LCDM

    k_fs = (0.08/np.sqrt(1.0+z)) * (m/0.1) * h
    q = 5.0*k/k_fs 
    Delta_L = 0.6*f_nu 
    Delta_q = 1.6 

    
    R_nu = 1.0 + 0.5 * Delta_L *( np.tanh(np.log(q)/Delta_q)+1.0 ) 

    return R_LCDM*R_nu 
    
def GISDB2(m1,m2,k,z):
    h=0.7
    m=np.array([m1,m2])
    Onuh2=m/93.14
    Ocdh2=0.14
    Omh2=Ocdh2+Onuh2.sum()
    f_nu=Onuh2/Omh2


    alpha = 4.0 
    Delta_LCDM = 4.8e-3 
    k_eq = 0.015 * h
    R_LCDM = 1.0 + Delta_LCDM * np.tanh(alpha*k/k_eq)

    Delta_q = 1.6 
    k_fs = (0.08/np.sqrt(1.0+z)) * (m/0.1) * h
    Delta_L = 0.6*f_nu

    q1 = 5.0*k/k_fs[0]
    q2 = 5.0*k/k_fs[1]
    
    R_nu = 1.0 + 0.5 * Delta_L[0] *( np.tanh(np.log(q1)/Delta_q)+1.0 )  + 0.5 * Delta_L[1] *( np.tanh(np.log(q2)/Delta_q)+1.0 ) 
    return R_LCDM*R_nu 

def GISDB3(m1,m2,m3,k,z):
    h=0.7
    m=np.array([m1,m2,m3])
    Onuh2=m/93.14
    Ocdh2=0.14
    Omh2=Ocdh2+Onuh2.sum()
    f_nu=Onuh2/Omh2

    alpha = 4.0 
    Delta_LCDM = 4.8e-3 
    k_eq = 0.015 * h
    R_LCDM = 1.0 + Delta_LCDM * np.tanh(alpha*k/k_eq)

    Delta_q = 1.6 
    k_fs = (0.08/np.sqrt(1.0+z)) * (m/0.1) * h
    Delta_L = 0.6*f_nu

    q1 = 5.0*k/k_fs[0]
    q2 = 5.0*k/k_fs[1]
    q3 = 5.0*k/k_fs[2]
    
    R_nu = 1.0 + 0.5 * Delta_L[0] *( np.tanh(np.log(q1)/Delta_q)+1.0 )  + 0.5 * Delta_L[1] *( np.tanh(np.log(q2)/Delta_q)+1.0 ) + 0.5 * Delta_L[2] *( np.tanh(np.log(q3)/Delta_q)+1.0 ) 

    return R_LCDM*R_nu 

```
